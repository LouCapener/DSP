```python
# LIBRARIES

import spacy
```

```python
# Lemmatization = process of reducing diff forms of a word to one single form
# e.g. reducing builds, building or built to the lemma "build"

nlp = spacy.load("en_core_web_sm", disable = ['ner', 'parser'])
```

```python
def cleaning(doc):
    # Lemmatizes and removes stopwords
    # doc needs to be a spacy Doc object
    txt = [token.lemma_ for token in doc if not token.is_stop]
    # Word2Vec uses context words to learn the vector representation of a target word,
    # if a sentence is only one or two words long,
    # the benefit for the training is very small
    if len(txt) > 2:
        return ' '.join(txt)
```

```python
brief_cleaning = (re.sub("[^A-Za-z']+", ' ', str(row)).lower() for row in merged['Tool'])
```

```python
txt = [cleaning(doc) for doc in nlp.pipe(brief_cleaning)]
```
